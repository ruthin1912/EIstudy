**Virtual Classroom Manager Programming Exercise**

```python
class Classroom:
    def __init__(self, name):
        self.name = name
        self.students = []
        self.assignments = []

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

class Assignment:
    def __init__(self, details):
        self.details = details
        self.submitted = False

class VirtualClassroomManager:
    def __init__(self):
        self.classrooms = []
        self.students = []

    def add_classroom(self, class_name):
        classroom = Classroom(class_name)
        self.classrooms.append(classroom)
        print(f"Classroom '{class_name}' has been created.")

    def add_student(self, student_id, class_name):
        classroom = next((c for c in self.classrooms if c.name == class_name), None)
        if classroom:
            student = Student(student_id, f"Student-{student_id}")
            classroom.students.append(student)
            print(f"Student {student_id} has been enrolled in {class_name}.")
        else:
            print(f"Error: Classroom '{class_name}' not found.")

    def schedule_assignment(self, class_name, assignment_details):
        classroom = next((c for c in self.classrooms if c.name == class_name), None)
        if classroom:
            assignment = Assignment(assignment_details)
            classroom.assignments.append(assignment)
            print(f"Assignment for {class_name} has been scheduled.")
        else:
            print(f"Error: Classroom '{class_name}' not found.")

    def submit_assignment(self, student_id, class_name, assignment_details):
        classroom = next((c for c in self.classrooms if c.name == class_name), None)
        if classroom:
            student = next((s for s in classroom.students if s.student_id == student_id), None)
            if student:
                assignment = next((a for a in classroom.assignments if a.details == assignment_details), None)
                if assignment:
                    assignment.submitted = True
                    print(f"Assignment submitted by Student {student_id} in {class_name}.")
                else:
                    print(f"Error: Assignment '{assignment_details}' not found.")
            else:
                print(f"Error: Student {student_id} not found in {class_name}.")
        else:
            print(f"Error: Classroom '{class_name}' not found.")

    def list_classrooms(self):
        print("\nList of Classrooms:")
        for classroom in self.classrooms:
            print(classroom.name)

    def list_students(self, class_name):
        classroom = next((c for c in self.classrooms if c.name == class_name), None)
        if classroom:
            print(f"\nList of Students in {class_name}:")
            for student in classroom.students:
                print(f"{student.name} (ID: {student.student_id})")
        else:
            print(f"Error: Classroom '{class_name}' not found.")

    def list_assignments(self, class_name, show_submitted=False):
        classroom = next((c for c in self.classrooms if c.name == class_name), None)
        if classroom:
            print(f"\nList of Assignments in {class_name}:")
            for assignment in classroom.assignments:
                if show_submitted or not assignment.submitted:
                    status = "Submitted" if assignment.submitted else "Not Submitted"
                    print(f"{assignment.details} - {status}")
        else:
            print(f"Error: Classroom '{class_name}' not found.")

    def overview(self):
        print("\nOverview:")
        print(f"Number of Classrooms: {len(self.classrooms)}")
        print(f"Number of Students: {sum(len(c.students) for c in self.classrooms)}")
        print(f"Number of Assignments: {sum(len(c.assignments) for c in self.classrooms)}")

if __name__ == "__main__":
    manager = VirtualClassroomManager()

    while True:
        print("\nVirtual Classroom Manager")
        print("1. Add Classroom")
        print("2. Add Student")
        print("3. Schedule Assignment")
        print("4. Submit Assignment")
        print("5. List Classrooms")
        print("6. List Students in a Classroom")
        print("7. List Submitted Assignments in a Classroom")
        print("8. Overview")
        print("9. Exit")

        choice = input("Enter your choice (1-10): ")

        if choice == "1":
            class_name = input("Enter the name of the classroom: ")
            manager.add_classroom(class_name)
        elif choice == "2":
            student_id = input("Enter the student ID: ")
            class_name = input("Enter the name of the classroom: ")
            manager.add_student(student_id, class_name)
        elif choice == "3":
            class_name = input("Enter the name of the classroom: ")
            assignment_details = input("Enter assignment details: ")
            manager.schedule_assignment(class_name, assignment_details)
        elif choice == "4":
            student_id = input("Enter the student ID: ")
            class_name = input("Enter the name of the classroom: ")
            assignment_details = input("Enter assignment details: ")
            manager.submit_assignment(student_id, class_name, assignment_details)
        elif choice == "5":
            manager.list_classrooms()
        elif choice == "6":
            class_name = input("Enter the name of the classroom: ")
            manager.list_students(class_name)
        elif choice == "7":
            class_name = input("Enter the name of the classroom: ")
            manager.list_assignments(class_name, show_submitted=True)
        elif choice == "8":
            manager.overview()
        elif choice == "9":
            print("Exiting Virtual Classroom Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```

**Personal To-Do List Manager Programming Exercise**

```python
from datetime import datetime

class Task:
    def __init__(self, description, due_date=None):
        self.description = description
        self.completed = False
        self.due_date = due_date

    def mark_completed(self):
        self.completed = True

    def __str__(self):
        status = "Completed" if self.completed else "Pending"
        task_str = f"{self.description} - {status}"
        if self.due_date:
            task_str += f", Due: {self.due_date.strftime('%Y-%m-%d')}"
        return task_str

class TaskBuilder:
    def __init__(self, description):
        self.task = Task(description)

    def with_due_date(self, due_date):
        self.task.due_date = due_date
        return self

    def build(self):
        return self.task

class TaskListManager:
    def __init__(self):
        self.tasks = []
        self.history = []

    def add_task(self, task):
        self.tasks.append(task)
        self.save_state()

    def mark_completed(self, description):
        for task in self.tasks:
            if task.description == description:
                task.mark_completed()
                self.save_state()
                return

    def delete_task(self, description):
        self.tasks = [task for task in self.tasks if task.description != description]
        self.save_state()

    def view_tasks(self, filter_option=None):
        if filter_option == "completed":
            filtered_tasks = [task for task in self.tasks if task.completed]
        elif filter_option == "pending":
            filtered_tasks = [task for task in self.tasks if not task.completed]
        else:
            filtered_tasks = self.tasks

        return filtered_tasks

    def save_state(self):
        self.history.append(list(self.tasks))

    def undo(self):
        if len(self.history) > 1:
            self.history.pop()
            self.tasks = self.history[-1]

    def display_tasks(self, tasks):
        if not tasks:
            print("No tasks found.")
        else:
            for task in tasks:
                print(task)

def display_menu():
    print("\nMenu:")
    print("1. Add Task")
    print("2. Mark Task as Completed")
    print("3. Delete Task")
    print("4. View Tasks")
    print("5. Undo")
    print("6. Exit")

if __name__ == "__main__":
    task_manager = TaskListManager()

    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == "1":
            description = input("Enter task description: ")
            due_date_str = input("Enter due date (format YYYY-MM-DD): ")
            due_date = datetime.strptime(due_date_str, "%Y-%m-%d") if due_date_str else None
            task = TaskBuilder(description).with_due_date(due_date).build()
            task_manager.add_task(task)
            print("Task added successfully.")

        elif choice == "2":
            description = input("Enter task description to mark as completed: ")
            task_manager.mark_completed(description)
            print("Task marked as completed.")

        elif choice == "3":
            description = input("Enter task description to delete: ")
            task_manager.delete_task(description)
            print("Task deleted.")

        elif choice == "4":
            filter_option = input("Enter filter option (all/completed/pending): ")
            tasks = task_manager.view_tasks(filter_option)
            task_manager.display_tasks(tasks)

        elif choice == "5":
            task_manager.undo()
            print("Undo successful.")

        elif choice == "6":
            print("Exiting the To-Do List Manager. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a valid option.")
```

**Personal To-Do List Manager Programming Exercise(With GUI)**

```python
import tkinter as tk
from tkinter import ttk

class Task:
    def __init__(self, description, due_date=None):
        self.description = description
        self.completed = False
        self.due_date = due_date

    def mark_completed(self):
        self.completed = True

    def mark_pending(self):
        self.completed = False

    def __str__(self):
        status = "Completed" if self.completed else "Pending"
        task_info = f"{self.description} - {status}"
        if self.due_date:
            task_info += f", Due: {self.due_date}"
        return task_info

class TaskManager:
    def __init__(self):
        self.tasks = []
        self.history = []

    def add_task(self, task):
        self.history.append(self.tasks.copy())
        self.tasks.append(task)

    def delete_task(self, task_description):
        self.history.append(self.tasks.copy())
        self.tasks = [task for task in self.tasks if task.description != task_description]

    def mark_completed(self, task_description):
        self.history.append(self.tasks.copy())
        for task in self.tasks:
            if task.description == task_description:
                task.mark_completed()

    def mark_pending(self, task_description):
        self.history.append(self.tasks.copy())
        for task in self.tasks:
            if task.description == task_description:
                task.mark_pending()

    def undo(self):
        if len(self.history) > 1:
            self.tasks = self.history.pop()

    def get_filtered_tasks(self, filter_option=None):
        if filter_option == "Show completed":
            return [task for task in self.tasks if task.completed]
        elif filter_option == "Show pending":
            return [task for task in self.tasks if not task.completed]
        else:
            return self.tasks

class ToDoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List Manager")

        self.task_manager = TaskManager()

        self.task_entry = ttk.Entry(self.root, width=40)
        self.due_date_entry = ttk.Entry(self.root, width=15)
        self.add_button = ttk.Button(self.root, text="Add Task", command=self.add_task)

        self.task_listbox = tk.Listbox(self.root, selectmode=tk.SINGLE, width=50)
        self.task_listbox.bind('<Double-Button-1>', self.toggle_completion)

        self.show_all_button = ttk.Button(self.root, text="Show All", command=lambda: self.display_tasks())
        self.show_completed_button = ttk.Button(self.root, text="Show Completed", command=lambda: self.display_tasks("Show completed"))
        self.show_pending_button = ttk.Button(self.root, text="Show Pending", command=lambda: self.display_tasks("Show pending"))

        self.undo_button = ttk.Button(self.root, text="Undo", command=self.undo)

        self.task_entry.grid(row=0, column=0, padx=5, pady=5)
        self.due_date_entry.grid(row=0, column=1, padx=5, pady=5)
        self.add_button.grid(row=0, column=2, padx=5, pady=5)

        self.task_listbox.grid(row=1, column=0, columnspan=3, padx=5, pady=5)

        self.show_all_button.grid(row=2, column=0, padx=5, pady=5)
        self.show_completed_button.grid(row=2, column=1, padx=5, pady=5)
        self.show_pending_button.grid(row=2, column=2, padx=5, pady=5)

        self.undo_button.grid(row=3, column=0, columnspan=3, pady=10)

    def add_task(self):
        task_description = self.task_entry.get()
        due_date = self.due_date_entry.get()
        new_task = Task(task_description, due_date)
        self.task_manager.add_task(new_task)
        self.display_tasks()
        self.clear_entry_fields()

    def display_tasks(self, filter_option=None):
        self.task_listbox.delete(0, tk.END)
        tasks = self.task_manager.get_filtered_tasks(filter_option)
        for task in tasks:
            self.task_listbox.insert(tk.END, str(task))

    def toggle_completion(self, event):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            selected_task = self.task_listbox.get(selected_index)
            task_description = selected_task.split(' - ')[0]
            if "Completed" in selected_task:
                self.task_manager.mark_pending(task_description)
            else:
                self.task_manager.mark_completed(task_description)
            self.display_tasks()

    def undo(self):
        self.task_manager.undo()
        self.display_tasks()

    def clear_entry_fields(self):
        self.task_entry.delete(0, tk.END)
        self.due_date_entry.delete(0, tk.END)

if __name__ == "__main__":
    root = tk.Tk()
    app = ToDoApp(root)
    root.mainloop()
```
